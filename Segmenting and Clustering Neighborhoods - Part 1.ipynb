# Download Toronto Neighboorhood Information from Wikipedia and format it accordingly

### Import data from wikipedia

```python
# Import packages
import pandas as pd
import numpy as np
```

### Import data from wikipedia and drop rows from dataframe where 'borough' is equal to 'Not assigned'

```python
# Import data from wikipedia for Canada postal codes and put it into dataframe 'df_CA'
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_CA=pd.read_html(url,header=0)[0]
df_CA
```

```python
# Neighborhoods with the same neighborhood were already sorted by postal code
# and combined on the wikipedia page
```

```python
# Drop Postal codes with boroughs that have empty values
df_CA.replace('Not assigned',np.NaN,inplace=True)
df_CA.dropna(subset=['Borough'],inplace=True)
df_CA
```

### Using '.shape' method to print out number of rows in dataframe

```python
df_rows=df_CA.shape[0]
print('Number of roles in dataset:',df_rows)
```
